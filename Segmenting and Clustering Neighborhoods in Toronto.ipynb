```python
#First I impport data from the url
import requests
from bs4 import BeautifulSoup
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

page = BeautifulSoup(url,'lxml')
```

```python
#making sure I get true webdata
print(page.prettify())
```

```python
#get all table data
all_tables=page.find_all("table")
all_tables
```

```python
#making dataframe
import pandas as pd
df = pd.read_html(str(all_tables))

neighborhood=pd.DataFrame(df[0])
```

```python
#print(neighborhood.query('Borough != "Not assigned"'))
clean_data =neighborhood.query('Borough != "Not assigned"')
clean_data
```

```python
#I had to install folium
!pip install folium
import folium # map rendering library
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
#Tronto's geographical data
address = 'Toronto, Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values

map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10.7)

# add markers to map

for lat, lng, borough, neighbourhood in zip(True_data ['Latitude'], True_data ['Longitude'],True_data ['Borough'], True_data['Neighborhood']):

    label  = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)

    folium.CircleMarker(

           [lat, lng],
           radius=5,
           popup=label,
           color='blue',
           fill=True,
           fill_color='#3186cc',
           fill_opacity=0.7).add_to(map_Toronto)
#show the map
map_Toronto
```
